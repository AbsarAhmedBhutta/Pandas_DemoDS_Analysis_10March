# . Which month saw the highest joining of the employees

```python
highest_joining = df['Month of Joining'].value_counts().sort_index().head(10)
sortedmean = highest_joining.sort_values(ascending=False)
sortedmean.head(1)
```

# 2. What is the mean salary is the state KS

```python
ks_rows = df.loc[df['State'] == 'KS']
ks_rows['Salary'].mean()
```

# 3. Which Region has highest mean salary

```python
mean_salaries = df.groupby('Region')['Salary'].mean()
sortedmean = mean_salaries.sort_values(ascending=False)
highest = sortedmean.head(1)

print(f"highest mean Salary of Region is: {highest}")
```

# 4. Which region has lowest mean salary

```python
mean_salaries = df.groupby('Region')['Salary'].mean()
sortedmean = mean_salaries.sort_values(ascending=True)
lowest = sortedmean.head(1)

print(f"lowest mean Salary of Region is: {lowest}")
```

# 5. Which City has highest mean weight

```python
mean_Weight = df.groupby('City')['Weight in Kgs.'].mean()
sortedmean = mean_Weight.sort_values(ascending=False)
highest = sortedmean.head(1)

print(f"highest mean weight of City: {highest}")
```

# 6. What is the difference between mean salary of Doctors and Professors

```python
ks_rows = df.loc[df['Name Prefix'] == 'Dr.']
drs_slary_mean = ks_rows['Salary'].mean()
ks_rows = df.loc[df['Name Prefix'] == 'Prof.']
prof_slary_mean = ks_rows['Salary'].mean()

dif = prof_slary_mean-drs_slary_mean
print(f"difference between mean salary of Doctors and Professors: {dif}")
```

# 7. How many female use yahoo

```python
num_female_yahoo1 = df[(df['E Mail'].str.contains('@yahoo\.com$')) & (df['Gender'] == 'F')].shape[0]
num_female_yahoo2 = df[(df['E Mail'].str.contains('@yahoo\.ca$')) & (df['Gender'] == 'F')].shape[0]
num_female_yahoo3 = df[(df['E Mail'].str.contains('@yahoo\.co$')) & (df['Gender'] == 'F')].shape[0]
num_female_yahoo4 = df[(df['E Mail'].str.contains('@yahoo\.co.in$')) & (df['Gender'] == 'F')].shape[0]
num_female_yahoo5 = df[(df['E Mail'].str.contains('@yahoo\.co.uk$')) & (df['Gender'] == 'F')].shape[0]

num_female_yahoo = num_female_yahoo1 + num_female_yahoo2 + num_female_yahoo3 + num_female_yahoo4 +num_female_yahoo5


print(f"Number of females who use Yahoo: {num_female_yahoo}")
```

# 8. How many Doctors use gmail

```python
num_doctors_yahoo = df[(df['E Mail'].str.contains('@gmail\.com$')) & (df['Name Prefix'] == 'Dr.')].shape[0]

print(f"Number of Dr.s who use gmail: {num_doctors_yahoo}")
```
